# Data Visualization of Population Growth

I gathered the population data from csdojo.io/data and then stored the csv data into a **dataframe**.
Next, I used the **series** data type to plot the graph. We can see that both countries population is increasing. However, it seems that Chinas population is increasing at a faster rate.

```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv("countries.csv")
```

```python
data
```

```python
us = data[data.country == 'United States']
```

```python
us
```

```python
china = data[data.country == "China"]
```

```python
china
```

```python
plt.plot(us.year, us.population / 10 ** 6)
plt.plot(china.year, china.population / 10 ** 6)
plt.title("Population Growth")
plt.xlabel("Year")
plt.ylabel("Population (millions)")
plt.legend(["United States", "China"])
plt.show()
```

```python
plt.plot(us.year, us.population / us.population.iloc[0] * 100)
plt.plot(china.year, china.population / china.population.iloc[0] * 100)
plt.title("Population Percentage Growth")
plt.xlabel("Year")
plt.ylabel("Population Growth (%)")
plt.legend(["United States", "China"])
plt.show()
```
